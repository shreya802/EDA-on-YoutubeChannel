### Data Cleaning

```python

df.head(20)

#subscribers column consist of null value display in -- format.
```

```python
df = df.replace('--', np.nan, regex =True)
```

```python
df.head(20)
```

```python
df.isnull().sum()

# counting total number of null value in repective column
```

```python
# calculating percentage of null value in columns

per_miss = df.isnull().sum() * 100 / len(df)
per_miss
```

```python
sns.heatmap(df.isnull())

# white line show the null values 
```

```python
df.dropna(axis = 0, inplace = True )

# drop the rows consisting of NAN values.
# axis = 0 means rows and axis = 1 means columns
```

```python
sns.heatmap(df.isnull())
```

First we will remove the string "th" from the rank column by using str and then we have to replace "," by using str.replace and also change datatype object into int by using astype

```python
# check data type of all columns

df.dtypes
```

```python
# convert data type of rank from object to int

df['Rank'] = df['Rank'].str[0:-2]
```

```python
df.head()
```

```python
df.tail()
```

A Regular Expression (or Regex) is a pattern (or filter) that describes a set of strings that matches the pattern. In other words, a regex accepts a certain set of strings and rejects the rest. It is a sequence of characters that specifies a match pattern in text.

str.replace for substring replacements on a single string column, and replace for any general replacement on one or more columns.

```python
# remove , from rank numbers

df['Rank']= df['Rank'].str.replace(',' , "")
```

```python
df.tail()
```

```python
# convert datatype from object to int64

df['Rank']= df['Rank'].astype('int')
```

```python
df.head()
```

```python
df.tail()
```

```python
df.dtypes
```

First of all we will change the datatype of video uploads column(object to int) by using astype() method and then we have change the datatype of susbriber column(object to int) by using astype() method.

```python
df['Video Uploads']= df['Video Uploads'].astype('int')
```

```python
df['Subscribers']= df['Subscribers'].astype('int')
```

```python
df.dtypes
```

As you see grade column is in string format so we have to change it into interger format. To convert into integer we can map string value to interger value by using map function.

```python
 df["Grade"].unique()
```

```python
df["Grade"]=df["Grade"].map({'A++ ':5, 'A+ ':4, 'A ':3, 'A- ':2, 'B+ ':1})
```

```python
df.dtypes
```

```python
 df["Grade"].unique()
```
